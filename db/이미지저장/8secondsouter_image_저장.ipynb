```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup



import os
import urllib.parse as rep
import urllib.request as req
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless")


# webdriver 설정(Chrome, Firefox 등) - 일반 모드
browser = webdriver.Chrome('chromedriver.exe')
```

```python
# 크롬 브라우저 내부 대기
browser.implicitly_wait(5)


# 페이지 이동
browser.get('https://www.ssfshop.com/8Seconds/ssfshop/list?dspCtgryNo=SFMA41A02&brandShopNo=BDMA07A01&brndShopId=8SBSS&currentPage='+str(1)+'&sortColumn=NEW_GOD_SEQ&etcCtgryNo=&leftBrandNM=8SECONDS_8SBSS&serviceType=DSP&smtFlterVal=&price=&benefit=&delivery=&lineId=&ctgrySectCd=GNRL_CTGRY&brndId=&sizeNM=&colorCd=&materNM=&cateViewOn=&cateNo=&fitPsbYn=N')
```

```python
# Header 정보 초기화
opener = req.build_opener()
# User-Agent 정보
opener.addheaders = [('User-agent', UserAgent().ie)]
# Header 정보 삽입
req.install_opener(opener)
```

```python
cnt=0


browser.get('https://www.ssfshop.com/8Seconds/Outerwear/list?dspCtgryNo=SFMA41A07&brandShopNo=BDMA07A01&brndShopId=8SBSS&currentPage='+str(1)+'&sortColumn=NEW_GOD_SEQ&etcCtgryNo=&leftBrandNM=8SECONDS_8SBSS&serviceType=DSP&smtFlterVal=&price=&benefit=&delivery=&lineId=&ctgrySectCd=GNRL_CTGRY&brndId=&sizeNM=&colorCd=&materNM=&cateViewOn=&cateNo=&fitPsbYn=N')
for i in range(1,61):
    WebDriverWait(browser, 2).until(EC.presence_of_element_located((By.CSS_SELECTOR, '#dspGood > li:nth-child('+str(i)+') > a > span > img'))).click() 
    time.sleep(5)
    soup = BeautifulSoup(browser.page_source, "html.parser")
    prod_list = soup.select('#content > section.detail > div.summary > div.gallery > div.lSSlideOuter > div.lSPager.lSGallery > ul')
    p_img=[]
    for v in prod_list:
        for idx in range(3):
            img= v.select('img')[idx]['src']
            p_img.append(img)
            print(img)
    print()    
    savePath = "C:/image/8_outer_dozen/"+"outer_"+str(cnt)+"/"
    try:
        if not (os.path.isdir(savePath)):
            os.makedirs(os.path.join(savePath))
    except OSError as e:
        print("folder creation failed!")
        print("folder name : {}".format(e.filename))
        raise RuntimeError('System Exit!')
    else:
        print('folder is created!')
    cnt+=1        
    for i in range(0,len(p_img)) :
        fullFileName = os.path.join(savePath, savePath + str(i) + '.jpg')
        req.urlretrieve(p_img[i], fullFileName)
   
    del soup
    browser.get('https://www.ssfshop.com/8Seconds/Outerwear/list?dspCtgryNo=SFMA41A07&brandShopNo=BDMA07A01&brndShopId=8SBSS&currentPage='+str(1)+'&sortColumn=NEW_GOD_SEQ&etcCtgryNo=&leftBrandNM=8SECONDS_8SBSS&serviceType=DSP&smtFlterVal=&price=&benefit=&delivery=&lineId=&ctgrySectCd=GNRL_CTGRY&brndId=&sizeNM=&colorCd=&materNM=&cateViewOn=&cateNo=&fitPsbYn=N')



    time.sleep(2)
```
