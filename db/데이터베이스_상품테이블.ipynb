```python
top=pd.read_csv("8_seconds_csv/top.csv")
top.drop(top.columns[top.columns.str.contains('unnamed',case = False)],axis = 1,inplace=True)


outer=pd.read_csv("8_seconds_csv/outer.csv")
outer.drop(outer.columns[outer.columns.str.contains('unnamed',case = False)],axis = 1,inplace=True)

bottom=pd.read_csv("8_seconds_csv/bottom.csv")
bottom.drop(bottom.columns[bottom.columns.str.contains('unnamed',case = False)],axis = 1,inplace=True)


acc=pd.read_csv("8_seconds_csv/acc.csv")
acc.drop(acc.columns[acc.columns.str.contains('unnamed',case = False)],axis = 1,inplace=True)
```

```python
from random import random
p_price_org=[]

for i in range(len(p_price)):
    org_price=p_price[i].replace("원","").replace(",","")
    try:
        n = random()
        if n==0:
            n=0.9
        org_price=int(org_price)
        org_price=int(org_price)*n
        org_price=round(org_price)
        p_price_org.append(org_price)
        
    except ValueError:
        p_price_org.append("0")        
        

        
```
